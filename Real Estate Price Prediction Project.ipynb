# Step 3 : Feature Engineering


Calculates the "Resident Size" by dividing the "Median price" by the "Median PSF" for each row

```python
df6['Resident Size'] = (df6['Median price'] / df6['Median PSF']).round().astype(int)

print(df6[['Resident Size']])
```

```python
df6.head(20)
```

Calculate bedrooms and bathrooms based on the size

```python
df6['Resident Size'] = pd.to_numeric(df6['Resident Size'], errors='coerce')

def calculate_bedrooms_and_bathrooms(size):
    if size <= 1400:
        return 3, 2
    elif 1400 < size <= 2500:
        return 4, 3
    else:
        return 5, 4

df6[['Bedrooms', 'Bathrooms']] = df6['Resident Size'].apply(lambda size: pd.Series(calculate_bedrooms_and_bathrooms(size)))
df6[['Project/township/building', 'Resident Size', 'Bedrooms', 'Bathrooms']]


df7 = df6

print(df7[['Project/township/building', 'Resident Size', 'Bedrooms', 'Bathrooms']])
```


Histograms and Normal Distribution Fits of Median Price by State

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Set up subplots based on the 'State' column
states = df7['State'].unique()
num_states = len(states)

fig, axes = plt.subplots(nrows=num_states, ncols=1, figsize=(10, 6*num_states))

for i, state in enumerate(states):
    # Filter data for the current state
    data_state = df7[df7['State'] == state]['Median price']

    # Plot the histogram for the current state
    sns.histplot(data_state, bins=20, kde=True, color='blue', ax=axes[i], label=f'Histogram - {state}')

    # Fit a normal distribution to the data
    mu, std = norm.fit(data_state)
    xmin, xmax = axes[i].get_xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    axes[i].plot(x, p, 'k', linewidth=2, label=f'Fit result: $\mu$ = {mu:.2f}, $\sigma$ = {std:.2f}')

    # Adding labels and title for each subplot
    axes[i].set_xlabel('Median Price')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'Histogram and Fit of Median Price - {state}')

    # Show legend for each subplot
    axes[i].legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

Removing Outliers from Median PSF Data by Property Type

```python
import pandas as pd
import numpy as np

def remove_pps_outlier(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('Type'):
        m = np.mean(subdf['Median PSF'])
        st = 2 * np.std(subdf['Median PSF'])
        reduced_df = subdf[(subdf['Median PSF'] > (m-st)) & (subdf['Median PSF'] <= (m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df8 = remove_pps_outlier(df7)
df8.shape
```

```python
median_psf_summary_by_state = df8.groupby('State')['Median PSF'].describe()

print(median_psf_summary_by_state)
```
